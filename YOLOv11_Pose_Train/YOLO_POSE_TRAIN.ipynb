TESTING: SALVA IMMAGINI CON I KEYPOINTS E LE CONNESSIONI PER VEDERE LA MASCHERA E SE PREDICE BENE ANCHE LA SEQUENZIALITÀ

```python
from ultralytics import YOLO
import os
import cv2
import numpy as np
from pathlib import Path

model_path = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/YOLOv11_Pose_Train/runs/pose/staffe_finale_senza_freeze_small/weights/best.pt"
final_model = YOLO(model_path)


test_images_path = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Training_Test_Val/images/test"
output_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/YOLOv11_Pose_Train/runs/final_test_visuals/test_small_con_linee"

os.makedirs(output_dir, exist_ok=True)

connections = [(i, i + 1) for i in range(6)] #connessioni sequenziali dal valore 0 al 6 (avendo come massimo nunmero di keypoints 7)

kp_color = (0, 255, 0) #verde per i keypoint
line_color = (255, 0, 0) #blu per le linee di connessione
conf_threshold_kp = 0.5 #un keypoint viene disegnato solo se la sua confidenza è maggiore di questa soglia

print(f"\nInizio la generazione delle predizioni visive (solo keypoint e linee) sulle immagini del test set da: {test_images_path}")

image_paths = list(Path(test_images_path).glob("*.[jp][pn]g"))

for img_path in image_paths:
    img = cv2.imread(str(img_path))
    if img is None:
        print(f"Attenzione: Impossibile caricare l'immagine {img_path}. Saltando.")
        continue

    h_orig, w_orig, _ = img.shape #risoluzione originale

    results = final_model.predict(source=str(img_path), conf=0.5, imgsz=640, verbose=False) #viene fatto il predict

    result = results[0]

    base_size = min(h_orig, w_orig)
    kp_radius = max(1, int(base_size * 0.005))
    line_thickness = max(1, int(base_size * 0.002))

    #disegno di keypoints e connessioni
    if result.keypoints is not None:
        for kps in result.keypoints.data:

            visible_keypoints = []

            for i, (x_coord, y_coord, conf_kp) in enumerate(kps):

                x_coord_int = int(x_coord.item())
                y_coord_int = int(y_coord.item())
                conf_kp_float = conf_kp.item()

                if conf_kp_float > conf_threshold_kp:
                    visible_keypoints.append((x_coord_int, y_coord_int, i))  #keypoints visibili solo se sopra la soglia
                    cv2.circle(img, (x_coord_int, y_coord_int), kp_radius, kp_color, -1)

            #disegna le linee di connessione
            for start_idx, end_idx in connections:
                start_point = None
                end_point = None

                #trova le coordinate dei keypoint di inizio e fine nella lista dei visibili
                for vkp_x, vkp_y, vkp_orig_idx in visible_keypoints:
                    if vkp_orig_idx == start_idx:
                        start_point = (vkp_x, vkp_y)
                    if vkp_orig_idx == end_idx:
                        end_point = (vkp_x, vkp_y)

                #se entrambi i keypoint della connessione sono visibili, disegna la linea
                if start_point and end_point:
                    cv2.line(img, start_point, end_point, line_color, line_thickness, cv2.LINE_AA)


    out_path = os.path.join(output_dir, img_path.name)
    cv2.imwrite(out_path, img)

print(f"\n Immagini con keypoint e linee normalizzati salvate in:\n{output_dir}")
```

TESTING NUMERO DI KEYPOINTS PREDETTI PER OGNI IMMAGINE CON VALORE DI CONFIDENZA

Si nota che tutti i keypoints predetti con coordinata (0,0) hanno un valore di confidenza minore a 0.5 e sono sotto-soglia, perciò la gestione del numero di keypoints variabile funziona correttamente.

```python
import numpy as np
import os
import glob
from ultralytics import YOLO
import cv2

model_path = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/YOLOv11_Pose_Train/runs/pose/staffe_finale_senza_freeze_small/weights/best.pt"

test_images_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Training_Test_Val/images/test"

OBJECT_CONF_THRESHOLD = 0.5 #soglia di confidenza per i bounding box
IMGSZ_INFERENCE = 640

print(f"Caricamento del modello da: {model_path}")
model = YOLO(model_path)
print("Modello caricato con successo.")

print(f"\nInizio analisi dei keypoint predetti per le immagini in: {test_images_dir}")

image_files = glob.glob(os.path.join(test_images_dir, "*.jpg")) + \
              glob.glob(os.path.join(test_images_dir, "*.png")) + \
              glob.glob(os.path.join(test_images_dir, "*.jpeg"))

if not image_files:
    print(f"ERRORE: Nessuna immagine trovata nella directory: {test_images_dir}. Controlla il percorso.")
else:
    print(f"Trovate {len(image_files)} immagini nel test set.")

#ciclo su tutte le immagini di test
for img_path in image_files:
    img_name = os.path.basename(img_path)
    print(f"\n--- Analisi Immagine: {img_name} ---")
    results_list = model.predict(source=img_path, imgsz=IMGSZ_INFERENCE, conf=OBJECT_CONF_THRESHOLD, verbose=False)  #si esegue il predict

    if not results_list or not results_list[0].boxes:
        print(f"  Nessun oggetto rilevato sopra la soglia di confidenza {OBJECT_CONF_THRESHOLD}.")
        continue

    pred_result = results_list[0]

    num_instances = len(pred_result.boxes)

    if num_instances == 0:
        print(f"  Nessun oggetto rilevato sopra la soglia di confidenza {OBJECT_CONF_THRESHOLD}.")
        continue
    #itera sul numero di istanze nell'immagine
    for i in range(num_instances):
        box_conf = pred_result.boxes.conf[i].cpu().item()
        print(f"  Istanza {i+1} (Confidenza Oggetto: {box_conf:.4f}):")

        if pred_result.keypoints is None or pred_result.keypoints.xy.shape[0] <= i:
            print(f"    Nessun keypoint predetto per questa istanza (probabilmente confidenza keypoint troppo bassa o errore).")
            continue

        #estrai le coordinate e le confidenze dei keypoint per questa istanza
        instance_keypoints_coords = pred_result.keypoints.xy[i].cpu().numpy()
        instance_keypoints_conf = pred_result.keypoints.conf[i].cpu().numpy()

        num_kpts_pred = instance_keypoints_coords.shape[0]
        print(f"    Numero di Keypoint Predetti: {num_kpts_pred}")

        #stampa la confidenza per ogni singolo keypoint
        for kpt_idx in range(num_kpts_pred):
            kpt_coord = instance_keypoints_coords[kpt_idx]
            kpt_conf = instance_keypoints_conf[kpt_idx]
            print(f"      Keypoint {kpt_idx}: Coord [{kpt_coord[0]:.2f}, {kpt_coord[1]:.2f}], Confidenza: {kpt_conf:.4f}")

    print("-" * 40)

print("\n--- Analisi Completata ---")
```

TESTING: CALCOLO DELLA METRICA PCK

```python
import numpy as np
import os
import glob
from ultralytics import YOLO
import cv2
from tqdm import tqdm


def calculate_pck_for_instance(pred_keypoints, gt_keypoints_with_vis, gt_box_wh, threshold=0.1, gt_visibility_threshold=1): #calcola il valore di pck
    relevant_gt_kpts = [  #filtra i keypoints di ground truth con il flag di visibilità
        (idx, kpt[:2])
        for idx, kpt in enumerate(gt_keypoints_with_vis)
        if kpt[2] >= gt_visibility_threshold
    ]

    if not relevant_gt_kpts or pred_keypoints.shape[0] == 0:
        return 0.0

    object_scale = np.sqrt(gt_box_wh[0] * gt_box_wh[1]) if gt_box_wh[0] > 0 and gt_box_wh[1] > 0 else 1.0
    if object_scale < 1e-6:
        return 0.0

    correct = 0
    for idx, gt_xy in relevant_gt_kpts:
        if idx >= pred_keypoints.shape[0]:
            continue
        dist = np.linalg.norm(pred_keypoints[idx] - gt_xy) #calcola la distanza euclidea
        if dist / object_scale < threshold: #se la distanza normalizzata è inferiore alla soglia il keypointè corretto
            correct += 1

    return correct / len(relevant_gt_kpts)

def parse_yolo_label_file(file_path, img_width, img_height): #estrae le annotazioni di bounding box e keypoints
    annotations = []
    with open(file_path, 'r') as f:
        for line in f:
            parts = list(map(float, line.strip().split()))
            if not parts:
                continue
            class_id = int(parts[0])
            cx, cy, w, h = parts[1:5]
            x_center, y_center = cx * img_width, cy * img_height
            width, height = w * img_width, h * img_height
            x1, y1 = int(x_center - width / 2), int(y_center - height / 2)
            x2, y2 = int(x_center + width / 2), int(y_center + height / 2)

            keypoints_data = parts[5:]
            kpts_with_vis = [
                [keypoints_data[i] * img_width, keypoints_data[i+1] * img_height, keypoints_data[i+2]]
                for i in range(0, len(keypoints_data), 3) if (i + 2) < len(keypoints_data)
            ]

            annotations.append({
                'class_id': class_id,
                'box_pixel': np.array([x1, y1, x2, y2]),
                'box_wh_pixel': np.array([width, height]),
                'keypoints_pixel_with_vis': np.array(kpts_with_vis)
            })
    return annotations

def calculate_iou(boxA, boxB): #calcola l'IoU tra due bounding box

    xA, yA = max(boxA[0], boxB[0]), max(boxA[1], boxB[1])
    xB, yB = min(boxA[2], boxB[2]), min(boxA[3], boxB[3])
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    areaA = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    areaB = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
    return interArea / float(areaA + areaB - interArea) if (areaA + areaB - interArea) > 0 else 0.0


model_path = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/YOLOv11_Pose_Train/runs/pose/staffe_finale_senza_freeze_small/weights/best.pt"
test_images_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Training_Test_Val/images/test"
test_labels_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Training_Test_Val/labels/test"

PCK_THRESHOLDS = [0.05, 0.1, 0.2]  #soglie PCK
CONF_THRESHOLD = 0.5
IMGSZ_INFERENCE = 640
GT_VISIBILITY_THRESHOLD = 1 #soglia per considerare keypoint visibile

print(f"Caricamento del modello da: {model_path}")
model = YOLO(model_path)
print("Modello caricato con successo.")

all_pcks_per_instance = {thr: [] for thr in PCK_THRESHOLDS}
all_pcks_per_image = {thr: [] for thr in PCK_THRESHOLDS}
detailed_pck_results = {thr: [] for thr in PCK_THRESHOLDS}

image_files = glob.glob(os.path.join(test_images_dir, "*.jpg")) + \
              glob.glob(os.path.join(test_images_dir, "*.png")) + \
              glob.glob(os.path.join(test_images_dir, "*.jpeg"))

print(f"Trovate {len(image_files)} immagini nel test set.")

for img_path in tqdm(image_files, desc="Processing images"):  #loop sulle immagini e parsing
    img_name = os.path.basename(img_path)
    label_path = os.path.join(test_labels_dir, os.path.splitext(img_name)[0] + ".txt")
    if not os.path.exists(label_path):
        continue

    img = cv2.imread(img_path)
    if img is None:
        continue
    H, W = img.shape[:2]
    gt_annotations = parse_yolo_label_file(label_path, W, H)
    results = model.predict(source=img_path, imgsz=IMGSZ_INFERENCE, conf=CONF_THRESHOLD, verbose=False) #predizione del modello

    pck_by_threshold_per_image = {thr: [] for thr in PCK_THRESHOLDS}

    if results:
        pred_result = results[0]
        pred_boxes = pred_result.boxes.xyxy.cpu().numpy()
        pred_kpts = pred_result.keypoints.xy.cpu().numpy() if pred_result.keypoints is not None else []

        for gt in gt_annotations: #calcolo PCK per istanza
            gt_box = gt['box_pixel']
            gt_kpts = gt['keypoints_pixel_with_vis']
            gt_wh = gt['box_wh_pixel']

            if sum(k[2] >= GT_VISIBILITY_THRESHOLD for k in gt_kpts) == 0:
                continue

            best_iou, best_kpts = 0.0, None  #tra i bounding box predetti cerca quello con IoU massimo
            for j, p_box in enumerate(pred_boxes):
                iou = calculate_iou(gt_box, p_box)
                if iou > best_iou:
                    best_iou = iou
                    best_kpts = pred_kpts[j] if j < len(pred_kpts) else None

            if best_iou >= 0.5 and best_kpts is not None:
                for thr in PCK_THRESHOLDS:
                    pck = calculate_pck_for_instance(best_kpts, gt_kpts, gt_wh, threshold=thr, gt_visibility_threshold=GT_VISIBILITY_THRESHOLD)
                    pck_by_threshold_per_image[thr].append(pck)
                    all_pcks_per_instance[thr].append(pck)
            else:
                for thr in PCK_THRESHOLDS:
                    pck_by_threshold_per_image[thr].append(0.0)
                    all_pcks_per_instance[thr].append(0.0)
    else:
        for gt in gt_annotations:
            if sum(k[2] >= GT_VISIBILITY_THRESHOLD for k in gt['keypoints_pixel_with_vis']) > 0:
                for thr in PCK_THRESHOLDS:
                    pck_by_threshold_per_image[thr].append(0.0)
                    all_pcks_per_instance[thr].append(0.0)

    for thr in PCK_THRESHOLDS:
        if pck_by_threshold_per_image[thr]:
            mean_img_pck = np.mean(pck_by_threshold_per_image[thr])
            all_pcks_per_image[thr].append(mean_img_pck)
            detailed_pck_results[thr].append((img_name, mean_img_pck))
        else:
            detailed_pck_results[thr].append((img_name, 0.0))

for thr in PCK_THRESHOLDS:
    print(f"\n--- PCK@{thr} per immagine ---")
    for img_name, pck_val in sorted(detailed_pck_results[thr], key=lambda x: x[1], reverse=True):
        print(f"  Immagine: {img_name:<30} PCK@{thr}: {pck_val:.4f}")

print("\n--- Risultati PCK Complessivi ---")
for thr in PCK_THRESHOLDS:
    if all_pcks_per_instance[thr]:
        print(f"PCK@{thr} Medio (per Istanza): {np.mean(all_pcks_per_instance[thr]):.4f} su {len(all_pcks_per_instance[thr])} istanze")
    else:
        print(f"PCK@{thr} Medio (per Istanza): N/A")

    if all_pcks_per_image[thr]:
        print(f"PCK@{thr} Medio (per Immagine): {np.mean(all_pcks_per_image[thr]):.4f} su {len(all_pcks_per_image[thr])} immagini")
    else:
        print(f"PCK@{thr} Medio (per Immagine): N/A")
```

TESTING: CALCOLO DELLA METRICA OKS

```python
import os
import cv2
import numpy as np
from tqdm import tqdm

#parsing delle annotazioni
def parse_yolo_label_file(label_path, img_w, img_h):

    annotations = []
    with open(label_path, 'r') as f:
        for line in f.readlines():
            elems = line.strip().split()
            cls = int(elems[0])
            cx, cy, bw, bh = map(float, elems[1:5])
            keypoints_raw = list(map(float, elems[5:]))

            xmin = (cx - bw/2) * img_w
            xmax = (cx + bw/2) * img_w
            ymin = (cy - bh/2) * img_h
            ymax = (cy + bh/2) * img_h

            keypoints = []
            for i in range(0, len(keypoints_raw), 3):
                kx = keypoints_raw[i] * img_w
                ky = keypoints_raw[i+1] * img_h
                v = keypoints_raw[i+2]
                keypoints.append([kx, ky, v])

            annotations.append({
                'box_pixel': [xmin, ymin, xmax, ymax],
                'keypoints_pixel_with_vis': keypoints,
                'box_wh_pixel': [bw * img_w, bh * img_h]
            })
    return annotations

def calculate_iou(boxA, boxB): #funione che calcola l'IoU

    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    if interArea == 0:
        return 0.0
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    iou = interArea / float(boxAArea + boxBArea - interArea)
    return iou

def calculate_oks_per_keypoint(pred_kpts, gt_kpts_with_vis, gt_box_wh, sigma=0.1): #calcolo di OKS per ogni keypoint

    oks_list = []

    s = (gt_box_wh[0] * gt_box_wh[1]) #parametri s è l'area
    if s <= 0:
        s = 1e-6

    for i, (gx, gy, v) in enumerate(gt_kpts_with_vis):
        if v > 0:  #solo keypoint visibili
            if i < len(pred_kpts):
                dx = pred_kpts[i, 0] - gx
                dy = pred_kpts[i, 1] - gy

                denom = 2 * (sigma ** 2) * s
                if denom == 0:
                    oks = 0.0
                else:
                    e = (dx ** 2 + dy ** 2) / denom
                    oks = np.exp(-e)
            else:
                oks = 0.0
        else:
            oks = 0.0
        oks_list.append(oks)
    return oks_list

SIGMA = 0.1 #valore sigma
IOU_THRESHOLD = 0.2
IMGSZ_INFERENCE = 640
CONF_THRESHOLD = 0.5


model_path = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/YOLOv11_Pose_Train/runs/pose/staffe_finale_senza_freeze_small/weights/best.pt"
test_images_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Training_Test_Val/images/test"
test_labels_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Training_Test_Val/labels/test"

from ultralytics import YOLO

model = YOLO(model_path)

image_files = [os.path.join(test_images_dir, f) for f in os.listdir(test_images_dir)
               if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

oks_per_image = {}
all_oks_values = []

for img_path in tqdm(image_files, desc="Calcolo OKS per immagini"):
    img_name = os.path.basename(img_path)
    label_name = os.path.splitext(img_name)[0] + ".txt"
    label_path = os.path.join(test_labels_dir, label_name)

    if not os.path.exists(label_path):
        print(f"Label mancante per {img_name}, salto")
        continue

    img_cv = cv2.imread(img_path)
    if img_cv is None:
        print(f"Immagine non letta: {img_name}, salto")
        continue
    H_orig, W_orig, _ = img_cv.shape

    gt_annotations = parse_yolo_label_file(label_path, W_orig, H_orig)
    results_list = model.predict(source=img_path, imgsz=IMGSZ_INFERENCE, conf=CONF_THRESHOLD, verbose=False)

    if not results_list or not gt_annotations:
        print(f"Nessun risultato o GT per {img_name}, salto")
        continue

    pred_result = results_list[0]
    pred_boxes = pred_result.boxes.xyxy.cpu().numpy()
    pred_keypoints_all_instances = pred_result.keypoints.xy.cpu().numpy() if (pred_result.keypoints is not None and pred_result.keypoints.xy.shape[0] > 0) else []

    oks_per_instance = []

    for i, gt_ann in enumerate(gt_annotations):
        gt_box = gt_ann['box_pixel']
        gt_kpts_with_vis = gt_ann['keypoints_pixel_with_vis']
        gt_box_wh = gt_ann['box_wh_pixel']

        best_iou = 0.0
        best_pred_kpts_for_gt = None


        for j, p_box in enumerate(pred_boxes):
            iou = calculate_iou(gt_box, p_box)
            if iou > best_iou:

                if j < len(pred_keypoints_all_instances):
                    best_iou = iou
                    best_pred_kpts_for_gt = pred_keypoints_all_instances[j]
                else:

                    continue


        #aalcola OKS solo se una predizione valida è stata trovata e ha un IOU sufficiente
        if best_iou >= IOU_THRESHOLD and best_pred_kpts_for_gt is not None:
            oks_keypoints = calculate_oks_per_keypoint(
                best_pred_kpts_for_gt,
                gt_kpts_with_vis,
                gt_box_wh,
                sigma=SIGMA
            )
        else:
            oks_keypoints = [0.0 if v[2] > 0 else 0.0 for v in gt_kpts_with_vis]

        oks_per_instance.append(oks_keypoints)

        for idx_kpt, (gx, gy, v) in enumerate(gt_kpts_with_vis):
            if v > 0:
                all_oks_values.append(oks_keypoints[idx_kpt])


    oks_per_image[img_name] = oks_per_instance

    print(f"\nImmagine: {img_name}")
    for idx, oks_list in enumerate(oks_per_instance):
        oks_str = ", ".join([f"{oks_val:.3f}" for oks_val in oks_list])
        print(f"Istanza {idx}: OKS per keypoint: [{oks_str}]")

if len(all_oks_values) > 0:
    mean_total_oks = np.mean(all_oks_values)
    print(f"\n-------------------------------------------------")
    print(f"Media OKS Totale (su tutti i keypoint visibili): {mean_total_oks:.4f}")
    print(f"-------------------------------------------------")
else:
    print("\nNessun keypoint valido con OKS calcolato. Controllare i dati o le soglie.")
```
