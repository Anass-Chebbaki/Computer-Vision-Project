#**SCRIPT PER LA CONVERSIONE DELLE ESTENSIONI DELLE IMMAGINI**
Siccome alcune immagini sono state fornite in formati come ".bmp" allora con questo script si è in grado partendo da una directory di trasformarli tutti in JPEG tramite la libreria PILLOW in Python per garantire un'omogeneità del dataset sia in fase di annotazione che di fine-tuning.

```python
from PIL import Image
import os

input_folder = r"PathCartella"
output_folder = r"PathCartellaDestinazione"

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if file.endswith(".bmp"):
        img_path = os.path.join(input_folder, file)
        img = Image.open(img_path)
        output_path = os.path.join(output_folder, file.replace(".bmp", ".jpg"))
        img.convert("RGB").save(output_path, "JPEG")
        print(f"Converted: {file} → {output_path}")
```

#**SCRIPT PER LA RINOMINAZIONE AUTOMATICA DEI FILE**

Una volta sistemate le estensioni del file, il seguente script ha la funzione di rinominare le immagini contenute nel dataset e numerarle in ordine crescente.

1. Legge i file nella cartella input_folder

2. Se il file ha estensione valida, lo rinomina in:
staffe_001.jpg, staffe_002.png, ...

3. Mantiene l’estensione originale

4. Rinominazione ordinata in base all’ordine alfabetico attuale

```python
import os

# === CONFIGURAZIONE ===
input_folder = r"PathCartella"
nome_base = "staffe"

# === ESTENSIONI SUPPORTATE ===
estensioni_supportate = [".jpg", ".jpeg", ".png", ".bmp"]

# === RINOMINAZIONE ORDINATA ===
count = 1

# Ordina alfabeticamente per consistenza
for filename in sorted(os.listdir(input_folder)):
    ext = os.path.splitext(filename)[1].lower()

    if ext in estensioni_supportate:
        old_path = os.path.join(input_folder, filename)
        nuovo_nome = f"{nome_base}_{count:03d}{ext}"
        nuovo_path = os.path.join(input_folder, nuovo_nome)

        os.rename(old_path, nuovo_path)
        print(f"✅ {filename} → {nuovo_nome}")
        count += 1

print("\n✅ Rinomina completata.")
```

SCRIPT CHE FA RESIZE DELLE IMMAGINI DI TEST

```python
import os
import cv2
from tqdm import tqdm


input_images_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Training_Test_Val/images/test"
# Cartella dove verranno salvate le immagini ridimensionate
output_images_dir = "/content/drive/MyDrive/Computer_Vision_&_Deep_Learning/DATASET/Training_Test_Datasets/Images_test_resized"

# --- Dimensioni Desiderate ---
TARGET_SIZE = (640, 640) # (larghezza, altezza)

print(f"Inizio il ridimensionamento delle immagini da '{input_images_dir}' a '{output_images_dir}'")
print(f"Dimensione target: {TARGET_SIZE[0]}x{TARGET_SIZE[1]} pixel")

# Crea la cartella di output se non esiste
os.makedirs(output_images_dir, exist_ok=True)

# Ottieni la lista di tutti i file immagine nella cartella di input
image_files = [f for f in os.listdir(input_images_dir)
               if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]

if not image_files:
    print(f"Nessuna immagine trovata nella cartella: {input_images_dir}")
else:
    # Processa ogni immagine
    for image_name in tqdm(image_files, desc="Ridimensionamento immagini"):
        input_path = os.path.join(input_images_dir, image_name)
        output_path = os.path.join(output_images_dir, image_name)

        # Leggi l'immagine
        img = cv2.imread(input_path)

        if img is None:
            print(f"Attenzione: Impossibile leggere l'immagine '{image_name}'. Potrebbe essere corrotta o non un formato supportato. Salto.")
            continue

        # Ridimensiona l'immagine alla dimensione target
        # cv2.INTER_AREA è consigliato per il rimpicciolimento,
        # cv2.INTER_LINEAR o cv2.INTER_CUBIC per l'ingrandimento.
        # Per un resize generico, INTER_AREA è una buona scelta di default.
        resized_img = cv2.resize(img, TARGET_SIZE, interpolation=cv2.INTER_AREA)

        # Salva l'immagine ridimensionata
        cv2.imwrite(output_path, resized_img)

    print(f"\nRidimensionamento completato! Tutte le immagini sono state salvate in: {output_images_dir}")
```
